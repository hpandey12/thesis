```python
from PIL import Image, ImageShow

def crop_and_zoom(image_path, target_width, target_height, delta_x=0, delta_y=0, zoom=1.0):
    """
    Crops an image to a target size with an offset from the center and a zoom factor.

    Args:
        image_path (str): The path to the image file.
        target_width (int): The final desired width of the image.
        target_height (int): The final desired height of the image.
        delta_x (int, optional): Horizontal offset from the center. Defaults to 0.
        delta_y (int, optional): Vertical offset from the center. Defaults to 0.
        zoom (float, optional): Zoom factor. > 1 zooms in, < 1 zooms out. Defaults to 1.0.

    Returns:
        PIL.Image.Image: The cropped and resized image object, or None if an error occurs.
    """
    try:
        img = Image.open(image_path)
    except FileNotFoundError:
        print(f"Error: The file '{image_path}' was not found.")
        return None

    # 1. Calculate the size of the crop box based on the zoom factor
    # A higher zoom means a smaller box to cut from the original
    crop_width = target_width / zoom
    crop_height = target_height / zoom

    # Get original image dimensions
    original_width, original_height = img.size

    # 2. Calculate the center point with the user-defined offset
    center_x = original_width / 2 + delta_x
    center_y = original_height / 2 + delta_y

    # 3. Calculate the coordinates for the crop box
    left = center_x - crop_width / 2
    top = center_y - crop_height / 2
    right = center_x + crop_width / 2
    bottom = center_y + crop_height / 2
    
    # Ensure the crop box is within the image boundaries
    left = max(0, left)
    top = max(0, top)
    right = min(original_width, right)
    bottom = min(original_height, bottom)

    # 4. Crop the image using the calculated box
    cropped_img = img.crop((left, top, right, bottom))
    
    # 5. Resize the cropped image back to the target dimensions for the zoom effect
    # Using LANCZOS for high-quality resizing
    final_img = cropped_img.resize((target_width, target_height), Image.Resampling.LANCZOS)

    return final_img

# --- YOUR SETTINGS ---
IMAGE_PATH = r"C:\Users\harsh\Downloads\Mesh_Scene_3_mesh_00500.png"
IMAGE_PATH2 = r"C:\Users\harsh\Downloads\Mesh_Scene_3_mesh_00005.png"

# Desired final dimensions
DESIRED_WIDTH = 800
DESIRED_HEIGHT = 600

# Settings for the second image (offset and zoom)
OFFSET_X = 0
OFFSET_Y = 40
ZOOM_FACTOR = 1.2 # This will create a 20% zoom

# --- PROCESS THE IMAGES ---

# Process the first image (no offset, no zoom)
cropped_img_1 = crop_and_zoom(
    IMAGE_PATH,
    DESIRED_WIDTH,
    DESIRED_HEIGHT,
    delta_x=0,
    delta_y=0,
    zoom=1.05 # No zoom
)

# Process the second image (with offset and zoom)
cropped_img_2 = crop_and_zoom(
    IMAGE_PATH2,
    DESIRED_WIDTH,
    DESIRED_HEIGHT,
    delta_x=OFFSET_X,
    delta_y=OFFSET_Y,
    zoom=1.0 # Apply zoom
)

# --- DISPLAY OR SAVE THE RESULTS ---
if cropped_img_1:
    ImageShow.show(cropped_img_1, title="First Image (No Zoom)")
    # cropped_img_1.save("cropped_image_1.png")

if cropped_img_2:
    ImageShow.show(cropped_img_2, title="Second Image (With Zoom)")
    # cropped_img_2.save("cropped_image_2_zoomed.png")
```

```python
import os
from PIL import Image

def crop_and_zoom(image_path, target_width, target_height, delta_x=0, delta_y=0, zoom=1.0):
    """
    Crops an image to a target size with an offset from the center and a zoom factor.

    Args:
        image_path (str): The path to the image file.
        target_width (int): The final desired width of the image.
        target_height (int): The final desired height of the image.
        delta_x (int, optional): Horizontal offset from the center. Defaults to 0.
        delta_y (int, optional): Vertical offset from the center. Defaults to 0.
        zoom (float, optional): Zoom factor. > 1 zooms in, < 1 zooms out. Defaults to 1.0.

    Returns:
        PIL.Image.Image: The cropped and resized image object, or None if an error occurs.
    """
    try:
        img = Image.open(image_path)
    except FileNotFoundError:
        print(f"Error: Could not find file '{image_path}'")
        return None

    # Calculate the size of the crop box based on the zoom factor
    crop_width = target_width / zoom
    crop_height = target_height / zoom

    original_width, original_height = img.size
    
    # Calculate the center point with the user-defined offset
    center_x = original_width / 2 + delta_x
    center_y = original_height / 2 + delta_y

    # Calculate the coordinates for the crop box
    left = center_x - crop_width / 2
    top = center_y - crop_height / 2
    right = center_x + crop_width / 2
    bottom = center_y + crop_height / 2
    
    # Ensure the crop box is within the image boundaries
    left = max(0, left)
    top = max(0, top)
    right = min(original_width, right)
    bottom = min(original_height, bottom)

    # Crop the image and resize it back to the target dimensions for the zoom effect
    cropped_img = img.crop((left, top, right, bottom))
    final_img = cropped_img.resize((target_width, target_height), Image.Resampling.LANCZOS)

    return final_img

def process_image_sequence(config):
    """
    Processes a sequence of images, applying a constant crop, pan, and zoom effect.
    """
    input_dir = config['input_dir']
    output_dir = config['output_dir']
    
    # --- Ensure the output directory exists ---
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Created output directory: {output_dir}")

    # --- Loop through the image sequence ---
    for num in range(config['num_start'], config['num_end'] + 1, config['num_step']):
        # Format the number with 5 digits (e.g., 5 -> "00005")
        num_str = str(num).zfill(5)
        
        # Construct file paths
        filename = f"Mesh_Scene_3_mesh_{num_str}.png"
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename)

        print(f"Processing {filename}...")

        # --- Process the image with constant settings ---
        processed_image = crop_and_zoom(
            image_path=input_path,
            target_width=config['desired_width'],
            target_height=config['desired_height'],
            delta_x=config['offset_x'],
            delta_y=config['offset_y'],
            zoom=config['zoom_factor']
        )

        # --- Save the result ---
        if processed_image:
            processed_image.save(output_path)

    print("\nProcessing complete!")
    print(f"Cropped images are saved in: {output_dir}")


if __name__ == "__main__":
    # --- Main Configuration ---
    # Adjust these settings to your needs
    
    # 1. PATHS
    INPUT_DIRECTORY = r"D:\thesis\HARD_DATA\mesh"
    OUTPUT_DIRECTORY = r"D:\thesis\HARD_DATA\mesh_cropped"

    # 2. IMAGE SEQUENCE
    NUM_START = 405      # First number in the sequence
    NUM_END = 2925    # Last number in the sequence
    NUM_STEP = 5       # Increment

    # 3. CROP & RESIZE
    DESIRED_WIDTH = 800
    DESIRED_HEIGHT = 600

    # 4. STATIC TRANSFORM SETTINGS - Applied to ALL images
    ZOOM_FACTOR = 1.05     # 20% zoom
    OFFSET_X = 0          # Horizontal offset from center
    OFFSET_Y = 0         # Vertical offset from center

    # --- Package configuration and run processor ---
    config = {
        "input_dir": INPUT_DIRECTORY,
        "output_dir": OUTPUT_DIRECTORY,
        "num_start": NUM_START,
        "num_end": NUM_END,
        "num_step": NUM_STEP,
        "desired_width": DESIRED_WIDTH,
        "desired_height": DESIRED_HEIGHT,
        "zoom_factor": ZOOM_FACTOR,
        "offset_x": OFFSET_X,
        "offset_y": OFFSET_Y,
    }

    process_image_sequence(config)
```

```python
import os
import re

# --- Configuration ---
file_prefix = "temp_scenebig_temp_"
file_extension = ".png"
output_prefix = "frame_"
# -------------------

# Get the current directory
image_folder = r"D:\thesis\HARD_DATA\vid2"

# Find all matching image files
files_to_process = []
for filename in os.listdir(image_folder):
    if filename.startswith(file_prefix) and filename.endswith(file_extension):
        # Extract the number part of the filename
        num_str = filename.replace(file_prefix, "").replace(file_extension, "")
        try:
            # Convert the scientific notation string to a float
            value = float(num_str)
            files_to_process.append((value, filename))
        except ValueError:
            print(f"Could not parse number from: {filename}")

# Sort the list based on the numerical value (the first item in our tuple)
files_to_process.sort()

# Rename the files sequentially
print(f"Found {len(files_to_process)} images to rename.")
for i, (value, old_filename) in enumerate(files_to_process):
    # Create new filename with 5-digit zero-padding (e.g., 00001, 00002)
    new_filename = f"{output_prefix}{i+1:05d}{file_extension}"
    old_filepath = os.path.join(image_folder, old_filename)
    new_filepath = os.path.join(image_folder, new_filename)
    
    print(f"Renaming '{old_filename}' -> '{new_filename}'")
    os.rename(old_filepath, new_filepath)

print("\n✅ Renaming complete!")
```
